# Data Analysis

# Correlation Coefficients

```python
plt.rcParams['figure.facecolor'] = "#e6ecff"
plt.rcParams['axes.facecolor'] = "#e6ecff"
sns.heatmap(dc.corr())
dc.corr()
```

Writing and Reading score are related with maximum,so if students scores high in raeding then he can also score in reading.

Maths scores are least likely related so maths scores does not depend on other subjects

# Representation of Test scores

```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.violinplot(y='math score',data=dc,color='yellow',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading score',data=dc,color='red',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing score',data=dc,color='pink',linewidth=3)
plt.show()
```

In the above plots,Maximum scores of students in Maths,Reading,Writing are between 60-80, 45-80, 50-80 respectively. 

# Relation between race/ethnicity and test scores

```python
plt.rcParams['figure.facecolor'] = "#e6ecff"
plt.rcParams['axes.facecolor'] = "#e6ecff"
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='race/ethnicity',y='math score',data=dc,palette='seismic')
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='race/ethnicity',y='reading score',data=dc,palette='seismic')
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='race/ethnicity',y='writing score',data=dc,palette='seismic')
plt.title('WRITING SCORES')
plt.show()
```

In the graph, we can see that the students who belongs to race of group E are scoring higher in all subjects.

# Relation between test preparation and test scores

```python
plt.rcParams['axes.facecolor'] = "#ccffda"
plt.rcParams['figure.facecolor'] = "#ccffda"

plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course',y='math score',data=dc)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course',y='reading score',data=dc)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course',y='writing score',data=dc)
plt.title('WRITING SCORES')
plt.show()
```

From the graph,it shows that the students who completed the course scored higher in all subjects.

# Relation between gender and test scores

```python
plt.rcParams['figure.facecolor'] = "#ffffe6"
plt.rcParams['axes.facecolor'] = "#ffcccc"

plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
plt.title('MATH SCORES')
sns.barplot(x='gender',y='math score',data=dc)

plt.subplot(1, 3, 2)
plt.title('READING SCORES')
sns.barplot(x='gender',y='reading score',data=dc)

plt.subplot(1, 3, 3)
plt.title('WRITING SCORES')
sns.barplot(x='gender',y='writing score',data=dc)
plt.show()
```

From the first graph,we can say that boys scored higher scores in math than girls.

From the next two graphs, we can say girls scored higher than boys in reading and writing

# Relation between lunch and test scores

```python
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='lunch',y='math score',data=dc,palette='seismic')
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='lunch',y='reading score',data=dc,palette='seismic')
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='lunch',y='writing score',data=dc,palette='seismic')
plt.title('WRITING SCORES')
plt.show()
```

In the above graph, the students with lunch as standard scored higher

# Relation between test preparation course,gender and Test scores

```python
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course',y='math score',hue='gender',data=dc)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course',y='reading score',hue='gender',data=dc)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course',y='writing score',hue='gender',data=dc)
plt.title('WRITING SCORES')
plt.show()
```

In the above graph, both boys and girls scored higher who completed the course

# Relation between Parental level of education and test scores

```python
plt.rcParams['axes.facecolor'] = "#ccffda"
plt.rcParams['figure.facecolor'] ="#ccffda"
plt.figure(figsize=(22,16))
plt.subplot(2, 2, 1)
sns.barplot(x='parental level of education',y='math score',data=dc)
plt.title('MATH SCORES')
plt.subplot(2, 2, 2)
sns.barplot(x='parental level of education',y='reading score',data=dc)
plt.title('READING SCORES')
plt.subplot(2, 2, 3)
sns.barplot(x='parental level of education',y='writing score',data=dc)
plt.title('WRITING SCORES')
plt.show()
```

In the above three graphs, students scored higher whose parent's level of education is some college or the degree.
